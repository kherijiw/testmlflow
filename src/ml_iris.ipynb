```python
###########################################################
# 1. PREPARATION DE L'ENVIRONNEMENT
###########################################################

import os
import joblib
import numpy as np
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score, classification_report
from sklearn.ensemble import RandomForestClassifier
import pandas as pd

```

```python
###########################################################
# 2. ACCES / INGESTION DES DONNEES
###########################################################

def load_data():
    iris = load_iris()
    X = iris.data
    y = iris.target
    df = pd.DataFrame(X, columns=iris.feature_names)
    return X, y,df

```

```python
###########################################################
# 3. PREPARATION / NETTOYAGE DES DONNEES
###########################################################

def prepare_data(X, y, test_size=0.2):
    # split with stratification
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, 
        test_size=test_size, 
        random_state=42,
        stratify=y
    )
    return X_train, X_test, y_train, y_test
```

```python
###########################################################
# 4. ENTRAINEMENT DU MODELE AVEC PIPELINE
###########################################################

def train_model(X_train, y_train):
    # pipeline = standardisation + random forest
    pipeline = Pipeline([
        ("scaler", StandardScaler()),
        ("clf", RandomForestClassifier(n_estimators=200, random_state=42))
    ])

    pipeline.fit(X_train, y_train)
    return pipeline
```

```python

###########################################################
# 5. EVALUATION DU MODELE
###########################################################

def evaluate_model(model, X_test, y_test):
    preds = model.predict(X_test)
    acc = accuracy_score(y_test, preds)
    print("âœ… Accuracy:", acc)
    print("\nClassification Report:\n")
    print(classification_report(y_test, preds))
    return acc
```

```python
###########################################################
# 6. ENREGISTREMENT DU MODELE
###########################################################

def save_model(model, path="./model/iris_model.joblib"):
    joblib.dump(model, path)
    print(f"ðŸ“ ModÃ¨le enregistrÃ© sous : {path}")
```

```python
###########################################################
# 7. ORCHESTRATION DU PIPELINE (END-TO-END)
###########################################################

def pipeline_iris():
    print("ðŸš€ PIPELINE LANCE...\n")

    print("1ï¸âƒ£ Chargement des donnÃ©es...")
    X, y, iris = load_data()
    print(iris.head(5))

    print("2ï¸âƒ£ PrÃ©paration / Nettoyage...")
    X_train, X_test, y_train, y_test = prepare_data(X, y)

    print("3ï¸âƒ£ EntraÃ®nement du modÃ¨le...")
    model = train_model(X_train, y_train)

    print("4ï¸âƒ£ Ã‰valuation du modÃ¨le...")
    acc = evaluate_model(model, X_test, y_test)

    print("5ï¸âƒ£ Sauvegarde du modÃ¨le...")
    save_model(model)

    print("\nðŸŽ‰ Pipeline terminÃ© avec succÃ¨s !")
    return model, acc
```

```python
###########################################################
# 8. EXECUTION DU PIPELINE
###########################################################

if __name__ == "__main__":
    model, accuracy = pipeline_iris()
```
